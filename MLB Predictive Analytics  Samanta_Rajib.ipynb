#### 1. Explore Attendance Trends:

```python
# Average attendance by day of the week
avg_attendance_day = dataset1_csv.groupby('day_of_week')['attend'].mean().sort_values(ascending=False)
sns.barplot(x=avg_attendance_day.index, y=avg_attendance_day.values)
plt.title('Average Attendance by Day of the Week')
plt.show()
```

#### 2. Impact of Promotional Events:

```python
# Average attendance with and without bobblehead promotion
avg_attendance_bobblehead = dataset1_csv.groupby('bobblehead')['attend'].mean()
sns.barplot(x=avg_attendance_bobblehead.index, y=avg_attendance_bobblehead.values)
plt.title('Average Attendance with and without bobblehead Promotion')
plt.show()
```

```python
# Average attendance with and without fireworks promotion
avg_attendance_fireworks = dataset1_csv.groupby('fireworks')['attend'].mean()
sns.barplot(x=avg_attendance_fireworks.index, y=avg_attendance_fireworks.values)
plt.title('Average Attendance with and without fireworks Promotion')
plt.show()
```

```python
# Average attendance with and without cap promotion
avg_attendance_cap = dataset1_csv.groupby('cap')['attend'].mean()
sns.barplot(x=avg_attendance_cap.index, y=avg_attendance_cap.values)
plt.title('Average Attendance with and without cap Promotion')
plt.show()
```

```python
# Average attendance with and without shirt promotion
avg_attendance_shirt = dataset1_csv.groupby('shirt')['attend'].mean()
sns.barplot(x=avg_attendance_shirt.index, y=avg_attendance_shirt.values)
plt.title('Average Attendance with and without shirt Promotion')
plt.show()
```

#### 3. Weather Conditions:

```python
# Average attendance by temperature range
dataset1_csv['temp_range'] = pd.cut(dataset1_csv['temp'], bins=[50, 60, 70, 80, 90], labels=['50-60', '60-70', '70-80', '80-90'])
avg_attendance_temp = dataset1_csv.groupby('temp_range')['attend'].mean()
sns.barplot(x=avg_attendance_temp.index, y=avg_attendance_temp.values)
plt.title('Average Attendance by Temperature Range')
plt.show()
```

#### 4. Day/Night Games:

```python
# Average attendance for day and night games
avg_attendance_day_night = dataset1_csv.groupby('day_night')['attend'].mean()
sns.barplot(x=avg_attendance_day_night.index, y=avg_attendance_day_night.values)
plt.title('Average Attendance for Day and Night Games')
plt.show()
```

#### 5. Opponent Impact:

```python
# Average attendance by opponent
avg_attendance_opponent = dataset1_csv.groupby('opponent')['attend'].mean().sort_values(ascending=False)
sns.barplot(x=avg_attendance_opponent.index, y=avg_attendance_opponent.values)
plt.title('Average Attendance by Opponent')
plt.xticks(rotation=45)
plt.show()
```

### Summary :
1. Tuesday's attendance is more than other days,Monday's average attendance is lowest.
2. People like bobblehead & Shirt as promotion, whereas Fireworks does not have any impact on attendance.
3. Audience avoid cold Weather Conditions ,more people come when temperature is around 70-80.
4. There is not much attendance difference for day or night match.
5. Attendance is high while opponent are Angels,Mets,Nationals same time for Braves is low.
